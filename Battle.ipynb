```python
def getNearbyVenues(names, latitudes, longitudes, radius=500, LIMIT=100):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Mun', 
                  'Mun Latitude', 
                  'Mun Longitude', 
                  'Mun_op', 
                  'Mun_op Latitude', 
                  'Mun_op Longitude', 
                  'Mun_op Category']
    
    return(nearby_venues)
```

```python
# one hot encoding
mun_onehot = pd.get_dummies(location_mun[['Mun_op Category']], prefix="", prefix_sep="")

# add street column back to dataframe
mun_onehot['Mun'] = location_mun['Mun'] 

# move street column to the first column
fixed_columns = [mun_onehot.columns[-1]] + list(mun_onehot.columns[:-1])

#fixed_columns
mun_onehot = mun_onehot[fixed_columns]

mun_onehot.head()
```

```python
mun_grouped = mun_onehot.groupby('Mun').mean().reset_index()
mun_grouped
```

```python
mun_grouped.shape
```

```python
# What are the top 5 mun/facilities nearby profitable real estate investments?#

num_top_mun = 5

for hood in mun_grouped['Mun']:
    print("----"+hood+"----")
    temp = mun_grouped[mun_grouped['Mun'] == hood].T.reset_index()
    temp.columns = ['type','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_mun))
    print('\n')
```

```python
def return_most_common_type(row, num_top_mun):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_mun]
```

```python
num_top_mun = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top mun
columns = ['Mun']
for ind in np.arange(num_top_mun):
    try:
        columns.append('{}{} Most Common Type'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Type'.format(ind+1))
```

```python
# create a new dataframe
mun_sorted = pd.DataFrame(columns=columns)
mun_sorted['Mun'] = mun_grouped['Mun']

for ind in np.arange(mun_grouped.shape[0]):
    mun_sorted.iloc[ind, 1:] = return_most_common_type(mun_grouped.iloc[ind, :], num_top_mun)
```
